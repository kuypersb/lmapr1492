# Tâche 7 : diffractogramme

Dans la suite de cette tâche, les trois premiers pics du diffractogramme seront développés. A chaque pic sont associés un angle $2\theta$, une intensité, des indices hkl (plans correspondants) avec leur multiplicité ainsi qu'une distance. Notre matériau a une structure cubique comme nous avons pu le montrer dans le jupyter notebook lattice.ipynb. Pour un tel réseau, la distance interplan est donnée par $d_{hkl} = \frac{a}{\sqrt{h^2 + k^2 + l^2}}$. On peut remarquer que la présence des carrés dans cette formule implique un rôle symétrique de h, k et l. En effet, un plan (111) aura la même distance qu'un plan (1-11). C'est de là que vient la multiplicité des différents plans. Cette multiplicité indique donc le nombre de plan qui ont la même distance entre eux.



Il est possible que différents plans correspondent au même pic du diffractogramme selon la loi de Bragg. 

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from jupyter_jsmol.pymatgen import quick_view
```

```python
# Récupération de la structure
with MPRester("m4UKTJ8LxatvzyEV") as m:
    structure = m.get_structure_by_material_id("mp-505297")
```

```python
# Visualisation
view = quick_view(structure)
display(view)
```

```python
# Récupération des informations concernant le diffractogramme créé par un rayonnemnt X incident CuKα
xrd = XRDCalculator("CuKa1")
pattern = xrd.get_pattern(structure)
```

## Premier pic du diffractogramme

```python
print("Premier pic pour un rayonnement X incident CuKα de longueur d’onde: λ=1.54060 Å")
print("Angle 2Θ entre les 2 rayons :", pattern.x[0])
print("Intensité du premier pic :", pattern.y[0])
print("Indices hkl associés :", pattern.hkls[0])
print("Distance associée :", pattern.d_hkls[0])
```

## Deuxième pic du diffractogramme

```python
print("Deuxième pic pour un rayonnement X incident CuKα de longueur d’onde: λ=1.54060 Å")
print("Angle 2Θ entre les 2 rayons :", pattern.x[1])
print("Intensité du premier pic :", pattern.y[1])
print("Indices hkl associés :", pattern.hkls[1])
print("Distance associée :", pattern.d_hkls[1])
```

## Troisième pic du diffractogramme

```python
print("Troisième pic pour un rayonnement X incident CuKα de longueur d’onde: λ=1.54060 Å")
print("Angle 2Θ entre les 2 rayons :", pattern.x[2])
print("Intensité du premier pic :", pattern.y[2])
print("Indices hkl associés :", pattern.hkls[2])
print("Distance associée :", pattern.d_hkls[2])
```

## Diffractogramme

```python
print("Intensité des trois premiers pics en fonction des angles pour un rayonnement X incident CuKα de longueur d’onde: λ=1.54060 Å")
xrd.show_plot(structure, two_theta_range=(0,45))
```
